The rocks fall in the order shown above: first the - shape, then the + shape, and so on. Once the end of the list is reached, the same order repeats: the - shape falls first, sixth, 11th, 16th, etc.

The rocks don't spin, but they do get pushed around by jets of hot gas coming out of the walls themselves. A quick scan reveals the effect the jets of hot gas will have on the rocks as they fall (your puzzle input).

For example, suppose this was the jet pattern in your cave:
```
>>><<><>><<<>><>>><<<>>><<<><<<>><>><<>>
```
In jet patterns, < means a push to the left, while > means a push to the right. The pattern above means that the jets will push a falling rock right, then right, then right, then left, then left, then right, and so on. If the end of the list is reached, it repeats.

The tall, vertical chamber is exactly seven units wide. Each rock appears so that its left edge is two units away from the left wall and its bottom edge is three units above the highest rock in the room (or the floor, if there isn't one).

After a rock appears, it alternates between being pushed by a jet of hot gas one unit (in the direction indicated by the next symbol in the jet pattern) and then falling one unit down. If any movement would cause any part of the rock to move into the walls, floor, or a stopped rock, the movement instead does not occur. If a downward movement would have caused a falling rock to move into the floor or an already-fallen rock, the falling rock stops where it is (having landed on something) and a new rock immediately begins falling.

To prove to the elephants your simulation is accurate, they want to know how tall the tower will get after 2022 rocks have stopped (but before the 2023rd rock begins falling). In this example, the tower of rocks will be 3068 units tall.

How many units tall will the tower of rocks be after 2022 rocks have stopped falling?


````
// Data Types and Functions
public record struct Position {
    public long X;
    public long Y;

    public static Position operator+(Position position, Position other) {
        return new Position {
            X = position.X + other.X,
            Y = position.Y + other.Y,
        };
    }
    
    public static Position operator-(Position position, Position other) {
        return new Position {
            X = position.X - other.X,
            Y = position.Y - other.Y,
        };
    }
}

public record struct RockShape {
    public HashSet<Position> Points;

    public override string ToString() {
        var buffer = new StringBuilder();
        for(var y = Points.Min(p => p.Y); y <= Points.Max(p => p.Y); y++) {
            for(var x = Points.Min(p => p.X); x <= Points.Max(p => p.X); x++) {
                buffer.Append(Points.Contains(new Position { X = x, Y = y }) ? "#" : ".");
            }
            buffer.Append("\n");
        }
        return buffer.ToString();
    }
}

public RockShape ParseRockShape(IEnumerable<string> lines) {
    return new RockShape {
        Points = lines.Reverse()
            .SelectMany(
                (line, y) => line
                    .Select((c, x) => (Char: c, Pos: new Position { X = x, Y = y }))
                    .Where(item => item.Char == '#')
                    .Select(item => item.Pos)
            )
            .ToHashSet()
    };
}

public enum WindDirection {
    Left,
    Right,
}

public static Position Delta(WindDirection direction) {
    return direction switch {
        WindDirection.Left => new Position { X = -1, Y  = 0 },
        WindDirection.Right => new Position { X = 1, Y  = 0 },
        _ => throw new System.InvalidOperationException("Unknown wind direction"),
    };
}

public WindDirection ParseWindDirection(char c) {
    return c switch {
        '<' => WindDirection.Left,
        '>' => WindDirection.Right,
        _ => throw new System.InvalidOperationException("Unknown wind direction"),
    };
}
public List<WindDirection> ParseAllWindDirections(string input) {
    return input.Select(ParseWindDirection)
        .ToList();
}

public class State {
    public HashSet<Position> StableRocks { get; set; }
    public HashSet<Position> FallingRocks { get; set; }

    public void WindStep(WindDirection windDirection) {
        var newPositions = FallingRocks.Select(r => r + Delta(windDirection));
        if (newPositions.Any(p => p.X < 0 || p.X >= 7 || StableRocks.Contains(p)) == false) {
            FallingRocks = newPositions.ToHashSet();
        }
    }

    public bool FallStep() {
        var newPositions = FallingRocks.Select(r => r + new Position { X = 0, Y = -1});
        if (newPositions.Any(p => p.Y < 1 || StableRocks.Contains(p)) == false) {
            FallingRocks = newPositions.ToHashSet();
            return false;
        } else {
            return true;
        }
    }

    public override string ToString() {
        var buffer = new StringBuilder();
        var points = FallingRocks.Union(StableRocks);
        for(var y = points.Max(p => p.Y); y > 0 ; y--) {
            buffer.Append("|");
            foreach(var x in Enumerable.Range(0, 7)) {
                var position = new Position { X = x, Y = y };
                if (FallingRocks.Contains(position)) {
                    buffer.Append("@");
                } else if (StableRocks.Contains(position)) {
                    buffer.Append("#");
                } else {
                    buffer.Append(".");
                }
            }
            buffer.Append("|");
            buffer.Append("\n");
        }
        buffer.Append("+-------+");
        return buffer.ToString();
    }
}

static IEnumerable<T> RepeatForever<T>(this IEnumerable<T> items) {
    while(true) {
        foreach (var item in items) {
            yield return item;
        }
    }
}

public record struct SimulationResult {
    public int SimulationLength;
    public long FinalHeight;
    public List<long> IncrementalHeightGains;
}

SimulationResult RunSimulation(State state, IEnumerator<RockShape> rockShapes, IEnumerator<WindDirection> windDirections, int numRocks) {
    var maxY = 0L;
    var gainedHeights = new List<long>();
    for (var rockNum = 1; rockNum <= numRocks; rockNum++) {
        var offset = new Position {
            X = 2L,
            Y = maxY + 4L,
        };
        rockShapes.MoveNext();
        state.FallingRocks = rockShapes.Current.Points
            .Select(p => p + offset)
            .ToHashSet();

        var finishedFall = false;
        while (finishedFall == false) {
            // Console.WriteLine(state);
            windDirections.MoveNext();
            state.WindStep(windDirections.Current);
            finishedFall = state.FallStep();
        }

        var newMaxY = Math.Max(maxY, state.FallingRocks.Max(p => p.Y));
        gainedHeights.Add(newMaxY - maxY);
        maxY = newMaxY;
        state.StableRocks.UnionWith(state.FallingRocks);
    }
    return new SimulationResult {
        SimulationLength = numRocks,
        FinalHeight = maxY,
        IncrementalHeightGains = gainedHeights,
    };
}

public record struct Pattern {
    public long CurrentHeight;
    public int NumFallenRocks;
    public int PatternSize;
    public long HeightPerPattern;
    public List<long> PatternItems;

    public long FinalHeight(long numTotalRocks) {
        var remainingRocks = numTotalRocks - NumFallenRocks;
        return CurrentHeight
            + (remainingRocks / PatternSize) * HeightPerPattern
            + PatternItems.Take((int) (remainingRocks % (long) PatternSize)).Sum();
    }
}

Pattern? FindRepeatingPattern(SimulationResult simResult, int patternSize) {
    var section1 = simResult.IncrementalHeightGains.Skip(simResult.SimulationLength - patternSize * 2)
            .Take(patternSize);
    var section2 = simResult.IncrementalHeightGains.Skip(simResult.SimulationLength - patternSize)
        .Take(patternSize);
    if (section1.Zip(section2).Any(pair => pair.First != pair.Second)) {
        return null;
    }
    return new Pattern {
        CurrentHeight = simResult.FinalHeight,
        NumFallenRocks = simResult.SimulationLength,
        PatternSize = patternSize,
        HeightPerPattern = section1.Sum(),
        PatternItems = section1.ToList(),
    };
}
````

````
var rocksInput = System.IO.File.ReadAllText("rocks.txt");

var input = System.IO.File.ReadAllText("input.txt");
input
````

````
var rockShapes = rocksInput.Split("\n\n")
    .Select(lineGroup => lineGroup.Split("\n"))
    .Select(ParseRockShape)
    .RepeatForever()
    .GetEnumerator();

var windDirections = ParseAllWindDirections(input)
    .RepeatForever()
    .GetEnumerator();

var state = new State {
    FallingRocks = new HashSet<Position>(),
    StableRocks = new HashSet<Position>(),
};

RunSimulation(state, rockShapes, windDirections, 2022).FinalHeight
````

How tall will the tower be after 1000000000000 rocks have stopped?

````
var rockShapes = rocksInput.Split("\n\n")
    .Select(lineGroup => lineGroup.Split("\n"))
    .Select(ParseRockShape)
    .RepeatForever()
    .GetEnumerator();

var windDirections = ParseAllWindDirections(input)
    .RepeatForever()
    .GetEnumerator();

var state = new State {
    FallingRocks = new HashSet<Position>(),
    StableRocks = new HashSet<Position>(),
};

var simulationLength = 10_000;
var simulationResult = RunSimulation(state, rockShapes, windDirections, simulationLength);

Pattern pattern = (Pattern) Enumerable.Range(100, (simulationLength / 2) - 100)
    .Select(patternSize => FindRepeatingPattern(simulationResult, patternSize))
    .First(pattern => pattern != null);

pattern.FinalHeight(1_000_000_000_000)
````
