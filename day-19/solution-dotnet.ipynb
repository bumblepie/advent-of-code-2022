````
// Data types and functions
using System.Text.RegularExpressions;

public record struct RobotCost {
    public Dictionary<string, int> Costs;
}

var RobotCostRegex = new Regex(@"(\d+) ([a-z]+)");
public RobotCost ParseRobotCost(string input) {
    return new RobotCost {
        Costs = RobotCostRegex.Matches(input)
            .Select(match => match.Groups)
            .ToDictionary(
                matchGroups => matchGroups[2].Value,
                matchGroups => int.Parse(matchGroups[1].Value)
            ),
    };
}

var AllRobotCostsRegex = new Regex(@"Each (\w+) robot costs ((?>\d+ [a-z]+(?> and )?)+)\.");
public Dictionary<string, RobotCost> ParseAllRobotCosts(string input) {
    return AllRobotCostsRegex.Matches(input)
        .Select(match => match.Groups)
        .ToDictionary(
            groups => groups[1].Value,
            groups => ParseRobotCost(groups[2].Value)
        );
}

public record struct Blueprint {
    public int BlueprintNumber;
    public Dictionary<string, RobotCost> RobotCosts;
    Dictionary<string, int> PrecalculatedMaxRobotsNeeded;

    public Dictionary<string, int> MaxRobotsNeeded() {
        if (PrecalculatedMaxRobotsNeeded == null) {
            var robotCosts = RobotCosts;
            PrecalculatedMaxRobotsNeeded = ResourceTypes
                .ToDictionary(r => r, r => robotCosts.Values.Max(costs => costs.Costs.GetValueOrDefault(r)));
        }
        return PrecalculatedMaxRobotsNeeded;
    }
}

var BlueprintNumberRegex = new Regex(@"Blueprint (\d+).*");
public Blueprint ParseBlueprint(string input) {
    return new Blueprint {
        BlueprintNumber = int.Parse(BlueprintNumberRegex.Match(input).Groups[1].Value),
        RobotCosts = ParseAllRobotCosts(input),
    };
}

public static readonly List<string> ResourceTypes = new List<string> {
    "geode",
    "obsidian",
    "clay",
    "ore",
};

public record struct State {
    public int Timestep;
    public string History;
    public Dictionary<string, int> CurrentResources;
    public Dictionary<string, int> CurrentRobots;

    // Assumes we can buy a geode for free every turn until the end
    public float MaxPossibleGeodes(int maxSteps) {
        float remainingSteps = maxSteps - Timestep;
        return CurrentResources["geode"]
            + (CurrentRobots["geode"] * remainingSteps)
            + ((remainingSteps + 1) * (remainingSteps / 2));
    }

    public int? StepsToAffordRobotFor(string resource, Blueprint blueprint) {
        var cost = blueprint.RobotCosts[resource];
        var resources = CurrentResources;
        var currentRobots = CurrentRobots;
        if (cost.Costs.Keys.Any(resource => currentRobots[resource] == 0)) {
            return null;
        }
        return cost.Costs.Keys
            .Select(resource => (
                MissingAmount: Math.Max(0, cost.Costs[resource] - resources[resource]),
                NumRobots: currentRobots[resource])
            )
            .Max(resourceStats => (int) Math.Ceiling( (float) resourceStats.MissingAmount / (float) resourceStats.NumRobots ));
    }

    public State BuyRobotFor(string resource, Blueprint blueprint) {
        var resources = CurrentResources.ToDictionary(kvp => kvp.Key, kvp => kvp.Value);
        foreach (var cost in blueprint.RobotCosts[resource].Costs) {
            resources[cost.Key] -= cost.Value;
        }
        var robots = CurrentRobots.ToDictionary(kvp => kvp.Key, kvp => kvp.Value);
        robots[resource] += 1;
        return new State {
            History = History + $"Bought {resource} Robot at timestep {Timestep}\n",
            Timestep = Timestep,
            CurrentResources = resources,
            CurrentRobots = robots,
        };
    }

    public State Mine(int steps) {
        var resources = CurrentResources.ToDictionary(kvp => kvp.Key, kvp => kvp.Value);
        foreach (var resource in ResourceTypes) {
            resources[resource] += steps * CurrentRobots[resource];
        }
        return new State {
            History = History + $"Mined for {steps} steps\n",
            Timestep = Timestep + steps,
            CurrentResources = resources,
            CurrentRobots = CurrentRobots.ToDictionary(kvp => kvp.Key, kvp => kvp.Value),
        };
    }

    public IEnumerable<State> NextStates(Blueprint blueprint, int remainingTime) {
        var currentState = this;
        var maxRobotsNeeded = blueprint.MaxRobotsNeeded();
        return ResourceTypes
            .Where(resource => resource == "geode" || currentState.CurrentRobots[resource] < maxRobotsNeeded[resource])
            .Select(resource => (Resource: resource, Steps: currentState.StepsToAffordRobotFor(resource, blueprint)))
            .Where(resource => resource.Steps != null && resource.Steps < remainingTime)
            .Select(resource => currentState
                .Mine((int) resource.Steps + 1)
                .BuyRobotFor(resource.Resource, blueprint)
            )
            .Append(currentState.Mine(remainingTime));
    }
}

// Breadth first search based solution - works better than DFS as DFS greedily makes the first robot it can
public State MineMostGeodesPossibleBFS(State startingState, Blueprint blueprint, int maxSteps) {
    var statesExplored = 0;

    var statesToExplore = new Queue<State>();
    statesToExplore.Enqueue(startingState);
    var maxGeodes = 0;
    State? maxGeodesState = null;

    while (statesToExplore.TryDequeue(out var state)) {
        statesExplored ++;
        if (state.Timestep == maxSteps) {
            if (state.CurrentResources["geode"] >= maxGeodes) {
                maxGeodesState = state;
                maxGeodes = state.CurrentResources["geode"];
            }
        } else if (state.MaxPossibleGeodes(maxSteps) <= maxGeodes) {
            continue;
        } else {
            foreach (var nextState in state.NextStates(blueprint, maxSteps - state.Timestep)) {
                statesToExplore.Enqueue(nextState);
            }
        }
    }
    var mgs = (State) maxGeodesState;
    return mgs;
}

// A* based priority queue solution, seems to be the better than BFS and DFS as it can terminate as
// soon as it finds a terminal state (ie maximum time has been reached).
public State MineMostGeodesPossiblePQ(State startingState, Blueprint blueprint, int maxSteps) {
    var statesExplored = 0;

    var statesToExplore = new PriorityQueue<State, float>();
    statesToExplore.Enqueue(startingState, -1 * startingState.MaxPossibleGeodes(maxSteps));

    while (statesToExplore.TryDequeue(out var state, out var _)) {
        statesExplored ++;
        if (state.Timestep == maxSteps) {
            return state;
        }
        foreach (var nextState in state.NextStates(blueprint, maxSteps - state.Timestep)) {
            statesToExplore.Enqueue(nextState, -1 * nextState.MaxPossibleGeodes(maxSteps));
        }
    }
    throw new System.InvalidOperationException("No terminating state found");
}
````

````
var blueprints = input.Select(ParseBlueprint).ToList();
var startingState = new State {
    History = "",
    Timestep = 0,
    CurrentResources = ResourceTypes.ToDictionary(r => r, _ => 0),
    CurrentRobots = ResourceTypes.ToDictionary(r => r, r => r == "ore" ? 1 : 0),
};

blueprints
    .AsParallel()
    .Select(blueprint => MineMostGeodesPossibleBFS(startingState, blueprint, 24))
    .Select((s, index) => (index + 1) * s.CurrentResources["geode"])
    .Sum()
````
